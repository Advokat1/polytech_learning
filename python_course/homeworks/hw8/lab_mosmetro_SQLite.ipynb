```python
# Мигрировать все данные из mosmetro.json файла в базу данных mosmetro.db

import json
import sqlite3

def json_reader(file_obj):
    data = json.load(file_obj)
    
    states = set()
    stations = set()
    lines = set()
    regions = set()
    
    for item in data:
        states.add(item['Status'])
        stations.add(item['Station'])
        lines.add(item['Line'])
        regions.add(item['District']) 
    
    data_dict = {
        'lines': lines,
        'stations': stations,
        'states': states,
        'regions': regions
    }

    return data_dict


if __name__ == "__main__":
    with open("mos_metro.json", "r") as f_obj:
        data_dict = json_reader(f_obj)

db_name = 'mos_metro.db'

connection = sqlite3.connect(db_name)
cursor = connection.cursor()

for key in data_dict.keys():
    col = ''

    if key == 'states':
        col = 'State'
    else:
        col = 'Name'
    
    for item in data_dict[f"{key}"]:
        cursor.execute(f"INSERT INTO {key}({col}) VALUES ('{item}')")
        connection.commit()
```

```python
# Мигрировать все данные из mosmetro.json файла в базу данных mosmetro.db

import json
import sqlite3

def json_reader(file_obj):
    
    db_name = 'mos_metro.db'

    connection = sqlite3.connect(db_name)
    cursor = connection.cursor()
    
    data = json.load(file_obj)

    for item in data:
        state = item['Status']
        station = item['Station']
        line = item['Line']
        region = item['District']

        cursor.execute(f"INSERT INTO metro_info(Line, Station, Region, State) VALUES ('{line}', '{station}', '{region}', '{state}')")
        connection.commit()

if __name__ == "__main__":
    with open("mos_metro.json", "r") as f_obj:
        json_reader(f_obj)
```
