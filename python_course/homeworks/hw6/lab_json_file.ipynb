```python
# Создать json файл, в котором будут храниться только станции в названии которых несколько слов

import json


def json_reader(file_obj):
    data = json.load(file_obj)
    
    stations = {}

    for item in data:
        if ' ' in item['Station']:
            stations[f'{item["ID"]}'] = item['Station']
    
    with open("stations.json", "w") as write_file:
        json.dump(stations, write_file, indent=2, ensure_ascii=False)
    
    

if __name__ == "__main__":
    with open("mos_metro.json", "r") as f_obj:
        json_reader(f_obj)
```

```python
# Создать N json файлов с данными по каждой ветке

import json


def make_json(file_name, file_obj):
    line_data = []
    
    with open(file_obj, "r") as f_obj:
        data = json.load(f_obj)   

        for item in data:
            if item['Line'] == file_name:
                line_data.append(
                    {
                        'Station': item['Station'],
                        'Status': item['Status'],
                        'ID': item['ID']
                    }
                )
    with open(f"{file_name}.json", "w") as write_file:
        json.dump(line_data, write_file, indent=2, ensure_ascii=False)


def json_reader(file_obj):
    data = json.load(file_obj)
    
    branches = set()

    for item in data:
        branches.add(item['Line'])
    
    return branches
    

if __name__ == "__main__":
    with open("mos_metro.json", "r") as f_obj:
        branches = json_reader(f_obj)
    
    for item in branches:
        make_json(item, 'mos_metro.json')
```

```python
# вывести в консоль список веток и количество станций на каждой из них

import json


def json_reader(file_obj):
    data = json.load(file_obj)
    
    branches = {}
    keys = branches.keys()
    
    for item in data:
        if item['Line'] in keys:
            temp = branches.get(f'{item["Line"]}')
            temp += 1
            branches[f'{item["Line"]}'] = temp
        else:
            branches[f'{item["Line"]}'] = 1
    
    print(branches)
    

if __name__ == "__main__":
    with open("mos_metro.json", "r") as f_obj:
        branches = json_reader(f_obj)
```
