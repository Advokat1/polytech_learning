```python
# Вывести в консоль топ 10 букв алфавита, с которых начинается название городов

import csv


def csv_dict_reader(file_obj):
    
    alphabet_stat = {}

    reader = csv.DictReader(file_obj, delimiter=';')
    
    keys = alphabet_stat.keys()
    
    for line in reader:
        if line['Город'][0] in keys:
            temp = alphabet_stat.get(f"{line['Город'][0]}")
            temp += 1
            alphabet_stat[f"{line['Город'][0]}"] = temp
        else:
            alphabet_stat[f"{line['Город'][0]}"] = 1

    return alphabet_stat


if __name__ == "__main__":
    with open("cities_russia.csv") as f_obj:
        stat_dict = csv_dict_reader(f_obj)
    
    sorted_dict = {}
    sorted_keys = sorted(stat_dict, key=stat_dict.get, reverse=True)
    
    for w in sorted_keys:
        sorted_dict[w] = stat_dict[w]
    
    i = 0
    
    for key, value in sorted_dict.items():
        i += 1
        if i <= 10:
            print(f'{repr(key):5} - {value}')
```

```python
# Вывести в консоль топ 10 городов с длинными названиями

import csv


def csv_dict_reader(file_obj):
    
    names_stat = {}

    reader = csv.DictReader(file_obj, delimiter=';')
    
    for line in reader:
        names_stat[f'{line["Город"]}'] = len(line["Город"])

    return names_stat


if __name__ == "__main__":
    with open("cities_russia.csv") as f_obj:
        stat_dict = csv_dict_reader(f_obj)
    
    sorted_dict = {}
    sorted_keys = sorted(stat_dict, key=stat_dict.get, reverse=True)
    
    for w in sorted_keys:
        sorted_dict[w] = stat_dict[w]
    
    i = 0
    
    for key, value in sorted_dict.items():
        i += 1
        if i <= 10:
            print(f'{repr(key):5} - {value}')
```

```python
# Вывести в консоль топ 10 городов в названии которых больше всего согласных букв

import csv


def csv_dict_reader(file_obj):
    
    chars = ('б', 'в', 'г', 'д', 'ж', 'з', 'й', 'к', 'л', 'м', 'н', 'п', 'р', 'с', 'т', 'у', 'ф', 'х', 'ц', 'ч', 'ш', 'щ', 'ъ', 'ь')
    
    names_stat = {}

    reader = csv.DictReader(file_obj, delimiter=';')
    
    for line in reader:
        count = 0
        for char in line['Город']:
            if char.lower() in chars:
                count += 1
        
        names_stat[f'{line["Город"]}'] = count

    return names_stat


if __name__ == "__main__":
    with open("cities_russia.csv") as f_obj:
        stat_dict = csv_dict_reader(f_obj)
    
    sorted_dict = {}
    sorted_keys = sorted(stat_dict, key=stat_dict.get, reverse=True)
    
    for w in sorted_keys:
        sorted_dict[w] = stat_dict[w]
    
    i = 0
    
    for key, value in sorted_dict.items():
        i += 1
        if i <= 10:
            print(f'{repr(key):5} - {value}')
```

```python
# Вывести в консоль названия городов, длинна которых <= 4. При этом список необходимо отсортировать в обратном порядке

import csv


def csv_dict_reader(file_obj):

    names_stat = []

    reader = csv.DictReader(file_obj, delimiter=';')
    
    for line in reader:
        if len(line['Город']) <= 4:
            names_stat.append(line['Город'])

    return names_stat


if __name__ == "__main__":
    with open("cities_russia.csv") as f_obj:
        stat_list = csv_dict_reader(f_obj)
    
    print(sorted(stat_list, reverse=True))
```
