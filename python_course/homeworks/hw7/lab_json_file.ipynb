```python
# Вывести топ 5 веток с наименьшим количеством действующих

import json


def json_reader(file_obj):
    data = json.load(file_obj)
    
    lines_stat = {}
    keys = lines_stat.keys()

    for item in data:
        if item['Status'] == 'действует':
            if item['Line'] in keys:
                temp = lines_stat.get(f'{item["Line"]}')
                temp += 1
                lines_stat[f'{item["Line"]}'] = temp
            else:
                lines_stat[f'{item["Line"]}'] = 1
        else:
            continue
    
    return lines_stat
    

if __name__ == "__main__":
    with open("mos_metro.json", "r") as f_obj:
        stat_dict = json_reader(f_obj)
    
    sorted_dict = {}
    sorted_keys = sorted(stat_dict, key=stat_dict.get, reverse=False)
    
    for w in sorted_keys:
        sorted_dict[w] = stat_dict[w]
    
    i = 0
    
    for key, value in sorted_dict.items():
        i += 1
        if i <= 5:
            print(f'{repr(key):5} - {value}')
```

```python
# Вывести топ 10 районов с наибольшим количеством станций метро в порядке возрастания

import json


def json_reader(file_obj):
    data = json.load(file_obj)
    
    districts_stat = {}
    keys = districts_stat.keys()

    for item in data:
        if item['District'] in keys:
            temp = districts_stat.get(f'{item["District"]}')
            temp += 1
            districts_stat[f'{item["District"]}'] = temp
        else:
            districts_stat[f'{item["District"]}'] = 1
    
    return districts_stat
    

if __name__ == "__main__":
    with open("mos_metro.json", "r") as f_obj:
        stat_dict = json_reader(f_obj)
    
    sorted_dict = {}
    sorted_keys = sorted(stat_dict, key=stat_dict.get, reverse=True)
    
    for w in sorted_keys:
        sorted_dict[w] = stat_dict[w]
    
    i = 0
    
    for key, value in sorted_dict.items():
        i += 1
        if i <= 10:
            print(f'{repr(key):5} - {value}')
```

```python
# Вывести в консоль список всех статусов станций (уникальные записи)

import json


def json_reader(file_obj):
    data = json.load(file_obj)
    
    states = set()
    
    for item in data:
        states.add(item['Status'])    
    
    return states
    

if __name__ == "__main__":
    with open("mos_metro.json", "r") as f_obj:
        states = json_reader(f_obj)
    print(states)
```

```python
# Создать json файл с названиями станций, линий и их статусами

import json


def json_reader(file_obj):
    data = json.load(file_obj)
    
    new_data = []

    for item in data:
        new_data.append(
            {
                'Station': item['Station'],
                'Line': item['Line'],
                'Status': item['Status']
            }
        )
    
    with open("stations.json", "w") as write_file:
        json.dump(new_data, write_file, indent=2, ensure_ascii=False)
    
    

if __name__ == "__main__":
    with open("mos_metro.json", "r") as f_obj:
        json_reader(f_obj)
```

```python
# Создать json файл с названиями станций, линий и их статусами

import json
import csv

def json_reader(file_obj):
    data = json.load(file_obj)

    with open('json_to_csv.csv', 'w', newline='') as csv_output:
        writer = csv.writer(csv_output, delimiter=';', lineterminator='\r')
        writer.writerow(['Station', 'Line', 'Status', 'AdmArea', 'District', 'ID'])
        for item in data:
            writer.writerow(
                [
                    item['Station'], 
                    item['Line'], 
                    item['Status'], 
                    item['AdmArea'], 
                    item['District'], 
                    item['ID']
                ]
            )


if __name__ == "__main__":
    with open("mos_metro.json", "r") as f_obj:
        json_reader(f_obj)
```

```python
# Создать json файл с двумерной логикой формирования файла. Первый уровень - название ветки. 
# Второй - все станции и информация о них в указанной ветке

import json


def json_reader(file_obj):
    data = json.load(file_obj)
    
    new_data = {}

    keys = new_data.keys()

    for item in data:
        if item['Line'] in keys:
            temp = new_data.get(f'{item["Line"]}')
            temp.append(
                {
                    'Station': item['Station'],
                    'ID': item['ID'],
                    'Status': item['Status']
                }
            )
        else:
            new_data[f'{item["Line"]}'] = [
                {
                    'Station': item['Station'],
                    'ID': item['ID'],
                    'Status': item['Status']   
                }
            ]

    with open("lines_stations.json", "w") as write_file:
        json.dump(new_data, write_file, indent=2, ensure_ascii=False)
    
    

if __name__ == "__main__":
    with open("mos_metro.json", "r") as f_obj:
        json_reader(f_obj)
```
