```python
# Создать временный файл, в который добавить названия случайных 10 городов из файла cities.csv, 
# а затем из этого файла создать другой файл, но уже с этими отсортированными в обратном алфавитном порядке городами. 
# Результат вывести в консоль для проверки.

import tempfile
import csv
import random


def csv_dict_reader(file_obj):
    cities = []
    reader = csv.DictReader(file_obj, delimiter=';')
    
    for line in reader:
        cities.append(line['Город'])

    return cities

if __name__ == "__main__":
    with open("cities_russia.csv") as f_obj:
        cities = csv_dict_reader(f_obj)

    rand_cities = random.sample(cities, 10)

    print(f"Случайный набор городов: {rand_cities}")

    file_tmp_first = tempfile.TemporaryFile()
    file_tmp_first.write(bytes(" ".join(rand_cities), 'utf-8'))
    file_tmp_first.seek(0)
    
    result = file_tmp_first.read()
    file_tmp_first.close()

    result = result.decode('utf-8')
    
    print(f"Данные в первом файле: {result}")

    
    result = sorted(result.split(' '), reverse=True)\

    file_tmp_second = tempfile.TemporaryFile()
    file_tmp_second.write(bytes(" ".join(result), 'utf-8'))
    file_tmp_second.seek(0)
    result = file_tmp_second.read()
    file_tmp_second.close()

    result = result.decode('utf-8')
    print(f"Сортированный набор названий второго файла {result}")
```

```python
# Создать временный каталог, в котором создать N текстовых файлов, 
# имя файла должно содержать название города из файла cities.csv  (Например, "Москва.txt", "Тула.txt"), 
# проверить результат своей работы, подсчитать количество созданных файлов и распечатать в консоль для проверки.

import tempfile
import os
import time


def csv_dict_reader(file_obj):
    cities = []
    reader = csv.DictReader(file_obj, delimiter=';')
    
    for line in reader:
        cities.append(line['Город'])

    return cities

if __name__ == "__main__":
    os.chdir('ENTER YOUR DIR')
    with open("cities_russia.csv") as f_obj:
        cities = csv_dict_reader(f_obj)
    
    with tempfile.TemporaryDirectory() as tmp_dir_name:
        print('created temporary directory', tmp_dir_name)
        for item in cities:
            with open(f'{tmp_dir_name}\\{item}.txt', 'w') as file:
                file.write(item)
            print(f'Created {item}.txt file')
        os.chdir(tmp_dir_name)
        
        print(len(os.listdir()))
        
        # Время, что бы убедиться в создании директории и файлов внутри неё
        time.sleep(25)
        os.chdir('ENTER YOUR DIR')
          
```
